```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("startup_funding.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Remarks'] , axis =1 , inplace=True)
```

```python
df.set_index('Sr No',inplace=True)
```

```python
df.rename(columns={'Date dd/mm/yyyy':'date',
                   'Startup Name':'startup',
                   'Industry Vertical':'vertical',
                   'SubVertical':'subvertical',
                   'City  Location':'city',
                   'Investors Name':'investors',
                   'InvestmentnType':'round',
                   'Amount in USD':'amount'

                   },inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['amount'] = df['amount'].fillna('0')
```

```python
df['amount'] = df['amount'].str.replace(',','')
df['amount'] = df['amount'].str.replace('undisclosed','0')
df['amount'] = df['amount'].str.replace('unknown','0')
df['amount'] = df['amount'].str.replace('Undisclosed','0')
```

```python
df = df[df['amount'].str.isdigit()]
```

```python
df['amount'] = df['amount'].astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
def to_inr(dollar):
    inr = dollar*90
    return inr/10000000

```

```python
df['amount']=df['amount'].apply(to_inr)
```

```python
df.head()
```

```python
df['date'] = df['date'].str.replace('05/072018','05/07/2018')
```

```python
df['date'] = pd.to_datetime(df['date'],errors='coerce')
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(subset = ['date', 'startup' , 'vertical', 'city' , 'investors' , 'round','amount'] ,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.to_csv('startup_cleaned.csv',index = False)
```
